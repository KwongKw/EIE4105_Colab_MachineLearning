```python
import scipy.io
noisy_train_digits = scipy.io.loadmat('../data/noisy_train_digits.mat')
```

```python
import numpy as np
for i in range(10):
    print('For digit ',i)
    y1 = noisy_train_digits['trainData'][i][0][:, (0, 399)]
    y2 = noisy_train_digits['trainData'][i][0][:, (399, 400)]
    print(y1.shape)
    print(y2.shape)
    C1 = np.cov(y1, rowvar=False, bias=True)
    print(C1)
    C2 = np.cov(y2, rowvar=False, bias=True)
    print(C2)
```

#Covariance matrices of images

```python
import numpy as np
import matplotlib.pyplot as plt

def show_cov(data):
    fig, ax = plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True, )
    ax = ax.flatten()
    for i in range(10):
        X = data['trainData'][i][0][:]
        C = np.cov(X, rowvar=False, bias=True)
        img = ax[i].imshow(C, interpolation='nearest')
        ax[i].title.set_text(f'Digit {i}')
    ax[0].set_xticks([])
    ax[0].set_yticks([])
    plt.tight_layout()
    plt.show()

show_cov(noisy_train_digits)
```
