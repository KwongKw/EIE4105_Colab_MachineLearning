```python
import scipy.io

# Load dataset
noisy_train_digits = scipy.io.loadmat('../data/noisy_train_digits.mat')
print(type(noisy_train_digits))
```

```python
# Show the digits
%matplotlib inline
import matplotlib.pyplot as plt

def show(data):
    fig, ax = plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True, )
    ax = ax.flatten()
    for i in range(10):
        img = data['trainData'][i][0][0].reshape(28, 28)  # Convert 784-dim vector to 28x28 image
        img = ax[i].imshow(img, interpolation='nearest')
    ax[0].set_xticks([])
    ax[0].set_yticks([])
    plt.tight_layout()
    plt.show()

show(noisy_train_digits)
```

# Show histogram

```python
def show_histogram(data):
    fig, axs = plt.subplots(5,2)
    fig.suptitle('histogram of the 400-th pixel of all digits')
    for i in range(10):
        x400 = data['trainData'][i][0][:, 399]    # 400-th pixel of of Digit i 
        n, bins, patches = axs[i % 5,i // 5].hist(x400, bins=256, density=True, facecolor='blue', alpha=0.75)
        axs[i % 5,i // 5].set_title(f'Digit {i}')
    fig.tight_layout()
    plt.show()
        
show_histogram(noisy_train_digits)
```
